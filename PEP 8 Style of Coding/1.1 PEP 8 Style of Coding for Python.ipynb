### **1.1.b Naming Style:**

The names we use when creating these labels need to follow a few rules:

1. Names can not start with a number.

2. There can be no spaces in the name, use _ instead.

3. Can't use any of these symbols :'",<>/?|\()!@#$%^&*~-+

4. Avoid using extra zero '0' before (prefix) numbers. 
   0.001 and 0+0j i.e. prefix is not an issue for decimal and complex number.
   
5. Multiple variable assignment
   x,y,z=100,200,300 (correct)
   x,y,z=100         (incorrect)
   
5. It's considered best practice (PEP8) that names are lowercase.

6. Avoid using the characters 'l' (lowercase letter el), 'O' (uppercase letter oh), 
   or 'I' (uppercase letter eye) as single character variable names.
   
7. Avoid using words that have special meaning in Python like "list" and "str"

### 1. Indentation

- A piece of code or a block of code which can be any loop, body of a function starts with indentation and ends in the alignment of first line (i.e., unindented line). There can be range or amount indentation which a user can follow throught the project is upto user. Generally, 4 spacebar spaces which equivalent to 1 tab space is consistently used in the industry. 

- The use of indentation make it looks clean and spacious which is ultimately easy to read for any other person or developer who haven't wrote that. 

**Note:** A coder can always ignore the indentation by continuation in line, but it is always preferred as to follow indentation is good habit of a developer which make the code more readable.

```python
# Consider a example to understand the indentation
for x in range(1,20):
    print(x)
    if x == 7:
        break
```

### 3. Comments in Python

- The most essential part of coding is comments. To understand what is happening in the code or a function or an expression, comments can be used to describe that. So, there could not be much time waste in understanding the source code.

- Now a days, developers are deveoping lot of projects and one can forget some part of code after a several time period. The short description of happening can be useful at that time. Comments are always high-yielding for beginners.

- In above examples, I have written some comments saying which is the correct way and which is the wrong way. It is helping the reader understand that what I am trying to explain in that part.

```python
# this is a loop
for i in range(51):           # initiating a loop
    if i > 48:                # if conditon statement
        print(str(i)+" Yes")  # this is an expression
    else:                     # else condition statement
        pass                  # ignoring the operation for that specific value of which does not satisfy the condition.
```

**Docstrings**
- Docstrings are documentation strings which are string literals and they appear in a fucntion as decription of function.

```python
# include an optional docstring to describe the effect of a function
def square_this(x):
    """Return the square of a number."""
    return x**2
print(square_this.__doc__)    # .__doc__  returns the docstring
```
